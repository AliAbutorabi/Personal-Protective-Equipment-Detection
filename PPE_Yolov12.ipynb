```python
%%writefile train_yolov12.sh
#!/bin/bash
source /usr/local/etc/profile.d/conda.sh
conda activate yolov12
cd yolov12

export MPLBACKEND=agg

python -c "
import matplotlib
matplotlib.use('agg')

from ultralytics import YOLO
model = YOLO('yolov12x.yaml')
model.train(
  data='/content/dataset/data.yaml',  # ЫҢШ§ Щ…ШіЫҢШұ ЩҒШ§ЫҢЩ„ Roboflow ШҙЩ…Ш§
  epochs=50,
  batch=8,
  imgsz=640,
  scale=0.5,
  mosaic=1.0,
  mixup=0.05,
  copy_paste=0.15,
  device='0'
)
"
```
